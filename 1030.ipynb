```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/gsy2.jpg')
print(img.shape)
glmg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(glmg)
```

```python
outlmg = cv2.blur(glmg,(7,7))
cv2_imshow(outlmg)
```

```python
outlmg = cv2.GaussianBlur(glmg,(7,7),5)
cv2_imshow(outlmg)
```

```python
outlmg = cv2.medianBlur(glmg,7)
cv2_imshow(outlmg)
```

```python
outlmage = glmg.copy()
blurlmage = cv2.medianBlur(outlmage, 7)

x = 200
y = 200
size = 15

for h in range(y, y+size):
  for w in range(x, x+size):
    outlmage[h][w] = blurlmage[h][w]

cv2_imshow(outlmage)
```

```python
outlmg = cv2.GaussianBlur(glmg,(7,7), 1)
cv2_imshow(outlmg)
```

```python
edgelmage = cv2.addWeighted(glmg, 1, outlmg, -1, 0)
cv2_imshow(edgelmage)
```

```python
unsharpmage = cv2.addWeighted(glmg, 1, edgelmage, 1, 0)
cv2_imshow(unsharpmage)
```
