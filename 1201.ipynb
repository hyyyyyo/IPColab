```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/abcdef.bmp')
print(img.shape)
glmg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (glmg1, cmap= 'gray')
cv2_imshow(glmg1)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
numLabels, labels = cv2.connectedComponents(gray_img)
print(numLabels)
```

```python
print(labels.shape)
print(labels)
```

```python
label_H = np.uint8(179 * labels / np.max(labels))
print(label_H)
```

```python
bank = 255 * np.ones_like(label_H)
labelImg = cv2.merge((label_H, bank, bank))
labelImg = cv2.cvtColor(labelImg, cv2.COLOR_HSV2RGB)

cv2_imshow(labelImg)
```

```python
bank = 255 * np.ones_like(label_H)
labelImg = cv2.merge((label_H, bank, bank))
labelImg = cv2.cvtColor(labelImg, cv2.COLOR_HSV2BGR)
labelImg[label_H == 0] = 0
cv2_imshow(labelImg)
```
