```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/fcespnoise.bmp')
print(img.shape)
glmg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(glmg1)
```

```python
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(glmg1, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
kernel = np.ones((3,3), np.uint8)
opening = cv2.dilate(erosion, kernel, iterations=1)
cv2_imshow(opening)
```
